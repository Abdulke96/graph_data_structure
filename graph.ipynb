## <span style="color: blue;">Types of Graph</span>

### <span style="color: teal;">Directed Graph</span>
A **directed graph** (or digraph) is a graph where edges have a direction. This means that an edge from node A to node B is not the same as an edge from node B to node A. Directed graphs are used to represent one-way relationships.


### <span style="color: teal;">Undirected Graph</span>
An **undirected graph** is a graph where edges have no direction. The edge between node A and node B is identical to the edge between node B and node A. Undirected graphs are used to represent two-way relationships, such as friendship in social networks.


## <span style="color: darkred;">Generating Random Graph</span>

```python
import random
def generate_random_graph(num_of_node: int, num_edges: int, max_num_of_neigh: int, is_directed:bool=False) -> dict:
    graph = {node:set() for node in range(num_of_node)}
    edges_added = 0
    while edges_added < num_edges:
        u = random.randint(0, num_of_node - 1)
        v = random.randint(0, num_of_node - 1)
        if u != v and v not in graph[u]:
            graph[u].add(v)
            if not is_directed:
                graph[v].add(u)
            edges_added += 1
    return {node: list(neigh) for node, neigh in graph.items()}

```

# <span style="color: teal;">Graph Traversal</span>

## <span style="color: darkorange;">Breadth First Search (BFS)</span>

### <span style="color: #2980b9;">Overview</span>
BFS is a <strong style="color: #2980b9;">graph traversal algorithm</strong> that explores nodes level by level. It starts at the <strong style="color: #2980b9;">source node</strong> and explores all its neighbors before moving on to the neighbors' neighbors. This ensures that all nodes at the present depth level are explored before moving on to nodes at the next depth level.

### <span style="color: #27ae60;">Algorithm Details</span>
- **Initialization**: BFS starts by initializing a queue and marking the source node as visited. The source node is then added to the queue.
- **Processing**: Nodes are dequeued one by one. For each node, all unvisited neighbors are marked as visited and enqueued.
- **Termination**: The algorithm continues until the queue is empty, meaning all reachable nodes have been processed.

### <span style="color: #e74c3c;">Time Complexity</span>
The time complexity of BFS is <code style="color: #e74c3c;">O(V + E)</code>, where:
- **<span style="color: #e74c3c;">`V`</span>** is the number of <strong style="color: #e74c3c;">vertices</strong> (nodes) in the graph.
- **<span style="color: #e74c3c;">`E`</span>** is the number of <strong style="color: #e74c3c;">edges</strong> in the graph. 

This complexity arises because each vertex and edge is processed at most once.

### <span style="color: #3498db;">Applications</span>
BFS has various applications in graph theory and computer science:
- **Shortest Path**: Finds the shortest path in unweighted graphs.
- **Connected Components**: Identifies connected components in a graph.
- **Level Order Traversal**: Used in level order traversal of trees.
- **Web Crawlers**: Utilized in web crawlers to explore web pages.


### <span style="color: #2ecc71;">Implementation  </span>



```python
import random
def BFS(graph:dict, start_node:int)->dict:
    visited = set()
    queue = [start_node]
    visited.add(start_node)
    result = {}
    while queue:
        node = queue.pop(0)
        result[node] = []
        for neigh in graph[node]:
            if neigh not in visited:
                visited.add(neigh)
                queue.append(neigh)
                result[node].append(neigh)
    return result

```

# <span style="color: teal;">Graph Traversal</span>

## <span style="color: darkorange;">Depth First Search (DFS)</span>

### <span style="color: #2980b9;">Overview</span>
DFS is a <strong style="color: #2980b9;">graph traversal algorithm</strong> that explores as far as possible along each branch before backing up. It starts at the <strong style="color: #2980b9;">source node</strong> and explores each branch of the graph before moving to the next branch. This technique uses a <strong style="color: #2980b9;">stack</strong> data structure (or recursion) to keep track of the nodes to be explored.

### <span style="color: #27ae60;">Algorithm Details</span>
- **Initialization**: DFS starts by initializing a stack and marking the source node as visited. The source node is then pushed onto the stack.
- **Processing**: Nodes are popped from the stack one by one. For each node, all unvisited neighbors are marked as visited and pushed onto the stack.
- **Termination**: The algorithm continues until the stack is empty, meaning all reachable nodes have been explored.

### <span style="color: #e74c3c;">Time Complexity</span>
The time complexity of DFS is <code style="color: #e74c3c;">O(V + E)</code>, where:
- **<span style="color: #e74c3c;">`V`</span>** is the number of <strong style="color: #e74c3c;">vertices</strong> (nodes) in the graph.
- **<span style="color: #e74c3c;">`E`</span>** is the number of <strong style="color: #e74c3c;">edges</strong> in the graph.

This complexity arises because each vertex and edge is processed at most once.

### <span style="color: #3498db;">Applications</span>
DFS has several important applications:
- **Path Finding**: Can be used to find paths between nodes in a graph.
- **Topological Sorting**: Used in scheduling problems to order tasks.
- **Cycle Detection**: Helps in detecting cycles in a graph.
- **Connected Components**: Identifies connected components in a graph.

### <span style="color: #8e44ad;">Example</span>
Consider a graph where you need to explore all nodes starting from node A. DFS will:
1. Start from node A, explore as far as possible along each branch.
2. Backtrack when no further nodes are available and continue from the next unvisited node.
3. Continue this process until all nodes are visited.

Using DFS, you can thoroughly explore all possible paths in a graph, making it useful for various applications that require comprehensive exploration.

### <span style="color: #2ecc71;">Implementation  </span>


```python

def DFS(graph: dict, start_node: int) -> dict:
    visited = set()
    stack = [start_node]
    visited.add(start_node)
    result = {}
    while stack:
        node = stack.pop(-1)
        result[node] = []
        for neigh in graph[node]:
            if neigh not in visited:
                visited.add(neigh)
                stack.append(neigh)
                result[node].append(neigh)
    return result
```

# <span style="color: teal;">Edge Types in Graph Traversal</span>

## <span style="color: darkorange;">Tree Edge</span>
#### A <strong style="color: #2980b9;">tree edge</strong> is an edge that is part of the <strong style="color: #2980b9;">depth-first search tree</strong>.
#### It connects a node to one of its <strong style="color: #2980b9;">descendants</strong> in the DFS traversal.
#### Tree edges form the structure of the DFS tree and represent the exploration path.

## <span style="color: darkorange;">Back Edge</span>
#### A <strong style="color: #e74c3c;">back edge</strong> is an edge that connects a node to one of its <strong style="color: #e74c3c;">ancestors</strong> in the DFS traversal.
#### It forms a <strong style="color: #e74c3c;">cycle</strong> in the graph.
#### Back edges indicate the presence of cycles and are used in cycle detection algorithms.

## <span style="color: darkorange;">Forward Edge</span>
#### A <strong style="color: #3498db;">forward edge</strong> is an edge that connects a node to one of its <strong style="color: #3498db;">descendants</strong> in the DFS traversal.
#### It does not form a cycle in the graph.
#### Forward edges skip intermediate nodes and directly connect to further descendants.

## <span style="color: darkorange;">Cross Edge</span>
#### A <strong style="color: #8e44ad;">cross edge</strong> is an edge that connects two nodes that are neither ancestors nor descendants of each other in the DFS traversal.
#### It does not form a cycle in the graph.
#### Cross edges connect nodes across different branches or trees in the DFS forest.


```python
import graphplotlab.draw_edge_types as dat
import networkx as nx
from typing import Dict, List, Tuple

def classify_edges(G: nx.Graph) -> Dict[Tuple[int, int], str]:
    edge_colors: Dict[Tuple[int, int], str] = {}
    state: Dict[int, int] = {}
    parent: Dict[int, int] = {}
    discovery_time: Dict[int, int] = {}
    time: List[int] = [0]  # Tracks the time during DFS

    # Node states
    NOT_DISCOVERED, IN_PROCESS, DISCOVERED = 0, 1, 2

    def dfs(v: int) -> None:
        state[v] = IN_PROCESS
        discovery_time[v] = time[0]
        time[0] += 1

        for neighbor in G.neighbors(v):
            neighbor_state = state.get(neighbor, NOT_DISCOVERED)
            if neighbor_state == NOT_DISCOVERED:
                parent[neighbor] = v
                edge_colors[(v, neighbor)] = 'blue'  # Tree edge
                dfs(neighbor)
            elif neighbor_state == IN_PROCESS:
                edge_colors[(v, neighbor)] = 'gray'  # Back edge
            elif neighbor_state == DISCOVERED:
                if discovery_time[v] < discovery_time[neighbor]:
                    edge_colors[(v, neighbor)] = 'red'  # Forward edge
                else:
                    edge_colors[(v, neighbor)] = 'orange'  # Cross edge

        state[v] = DISCOVERED

    # Run DFS from every unvisited node to ensure all components are covered
    for node in G.nodes():
        if state.get(node, NOT_DISCOVERED) == NOT_DISCOVERED:
            dfs(node)

    return edge_colors
```

# <span style="color: teal;">Topological Sorting</span>

### <span style="color: darkorange;">Overview</span>
Topological sorting is an ordering of the nodes in a <strong style="color: #2980b9;">directed acyclic graph (DAG)</strong> such that for every directed edge \( uv \) from node \( u \) to node \( v \), node \( u \) comes before node \( v \) in the ordering.

### <span style="color: darkorange;">Applications</span>
Topological sorting is widely used for:
- **<span style="color: #16a085;">Task Scheduling:</span>** Organizing tasks based on their dependencies.
- **<span style="color: #27ae60;">Dependency Resolution:</span>** Managing dependencies in software builds or package installations.
- **<span style="color: #8e44ad;">Graph-related Problems:</span>** Solving problems involving directed acyclic graphs.

### <span style="color: darkorange;">Algorithms</span>
Topological sorting can be implemented using:
1. **<span style="color: #3498db;">Depth-First Search (DFS)</span>**
2. **<span style="color: #e67e22;">Breadth-First Search (BFS)</span>**

### <span style="color: darkorange;">Time Complexity</span>
The time complexity of topological sorting is \( O(V + E) \), where:
- \( V \) is the number of vertices in the graph.
- \( E \) is the number of edges in the graph.

### <span style="color: darkorange;">Constraints</span>
- **<span style="color: #e74c3c;">Directed Acyclic Graph (DAG):</span>** Topological sorting is only possible in DAGs.
- **<span style="color: #e74c3c;">Cycles:</span>** If the graph contains a cycle, topological sorting is not feasible.

### <span style="color: darkorange;">Example</span>
Consider a simple DAG with vertices and directed edges. A valid topological sort will list vertices in an order such that for every directed edge \( uv \), \( u \) precedes \( v \).


### <span style="color: darkred;"> Depth First Search (DFS) Topological Sortings</span>


```python
import graphplotlab.draw_sorted_nodes as dsn
from typing import Dict, List, Set, Optional

def dfs(graph: Dict[int, List[int]], node: int, visited: Set[int], stack: List[int], rec_stack: Set[int]) -> bool:
    visited.add(node)
    rec_stack.add(node)

    for neighbor in graph.get(node, []):
        if neighbor not in visited:
            if not dfs(graph, neighbor, visited, stack, rec_stack):
                return False
        elif neighbor in rec_stack:
            # Cycle detected
            return False

    rec_stack.remove(node)
    stack.append(node)
    return True


def dfs_topological_sort(graph: Dict[int, List[int]]) -> Optional[List[int]]:
    visited = set()
    stack = []
    rec_stack = set()

    for node in graph:
        if node not in visited:
            if not dfs(graph, node, visited, stack, rec_stack):
                return None
    return stack[::-1]  
```


### <span style="color: darkred;"> Breadth First Search (BFS) Topological Sortings or Khan's Algorithm</span>


```python
from collections import deque
from typing import Dict, List

def khan(graph: Dict[int, List[int]]) -> List[int]:
    
    in_degree = {node: 0 for node in graph}
   
    for node in graph:
        for neighbor in graph[node]:
            in_degree[neighbor] += 1
    
    topo_sort = []
    queue = deque() 
    
 
    for node in graph:
        if in_degree[node] == 0:
            queue.append(node)
    
    while queue:
        node = queue.popleft()
        topo_sort.append(node)
        for neighbor in graph[node]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)
    
    return topo_sort 


    
```

# <span style="color: teal;">Connected Components in Graph Theory</span>

### <span style="color: darkorange;">Connected Components</span>
Connected components in a graph are subgraphs where all nodes are connected to each other by edges. Each connected component is a <strong style="color: #2980b9;">maximal subset</strong> of vertices such that there is a path between every pair of vertices within this subset.

### <span style="color: darkorange;">Types of Graphs Based on Connected Components</span>

- **<span style="color: #16a085;">Connected Graph:</span>** A graph with only one connected component. In a connected graph, there is a path between every pair of vertices.
  
- **<span style="color: #e74c3c;">Disconnected Graph:</span>** A graph with more than one connected component. In a disconnected graph, not all vertices are reachable from every other vertex.

### <span style="color: darkorange;">Finding Connected Components</span>

Connected components can be identified using graph traversal algorithms. The most common methods are:

1. **<span style="color: #3498db;">Depth-First Search (DFS)</span>**: 
   - A traversal method that explores as far as possible along each branch before backtracking.
   - DFS can be used to find all nodes in a connected component starting from any node in that component.

2. **<span style="color: #e67e22;">Breadth-First Search (BFS)</span>**: 
   - A traversal method that explores all neighbor nodes at the present depth prior to moving on to nodes at the next depth level.
   - BFS can also be used to identify all nodes in a connected component by starting from any node in the component.

### <span style="color: darkorange;">Applications</span>

Connected components are crucial in various domains such as:
- **<span style="color: #8e44ad;">Social Networks:</span>** To find groups of users who are closely connected.
- **<span style="color: #27ae60;">Communication Networks:</span>** To identify groups of nodes (like routers or computers) that are fully interconnected.

### <span style="color: darkorange;">DFS IMPLEMENTATION</span>


```python
from typing import Dict, List, Set, Tuple

def dfs(graph: Dict[int, List[int]], node: int, visited: Set[int], component: Dict[int, List[int]]) -> None:

    visited.add(node)
    if node not in component:
        component[node] = []
    for neighbor in graph.get(node, []):
        if neighbor not in visited:
            component[node].append(neighbor)
            dfs(graph, neighbor, visited, component)

def dfs_connected_components(graph: Dict[int, List[int]]) -> List[Dict[int, List[int]]]:
   
    visited = set()
    components = []
    for node in graph:
        if node not in visited:
            component = {}
            dfs(graph, node, visited, component)
            components.append(component)
    return components
```

### <span style="color: darkorange;">BFS IMPLEMENTATION</span>

```python
from typing import Dict, List, Set

def bfs(node: int, graph: Dict[int, List[int]], visited: Set[int]) -> Dict[int, List[int]]:
    queue = [node]
    visited.add(node)
    component: Dict[int, List[int]] = {node: []}
    
    while queue:
        current_node = queue.pop(0)
        for neighbor in graph[current_node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
                if current_node not in component:
                    component[current_node] = []
                component[current_node].append(neighbor)
                if neighbor not in component:
                    component[neighbor] = []
                component[neighbor].append(current_node)
    
    return component

def bfs_connected_components(graph: Dict[int, List[int]]) -> List[Dict[int, List[int]]]:
    visited = set()
    components = []
    
    for node in graph:
        if node not in visited:
            component = bfs(node, graph, visited)
            components.append(component)
    
    return components
```

# <span style="color: teal;">Strongly Connected Components</span>

### <span style="color: darkorange;">Strongly Connected Components (SCCs)</span>
In a directed graph, a **Strongly Connected Component (SCC)** is a maximal subgraph where every pair of vertices is reachable from each other. In other words, for any two vertices `u` and `v` within an SCC, there is a path from `u` to `v` and a path from `v` to `u`.

### <span style="color: darkorange;">Characteristics of SCCs</span>
- **<span style="color: #2980b9;">Maximal Subgraph:</span>** An SCC is the largest subset of vertices such that every vertex is reachable from every other vertex in the subset.
- **<span style="color: #2980b9;">Mutual Reachability:</span>** In an SCC, if there is a path from vertex `u` to vertex `v`, there is also a path from `v` to `u`.

### <span style="color: darkorange;">Finding SCCs</span>

To find SCCs in a directed graph, we commonly use:
1. **<span style="color: #3498db;">Kosaraju's Algorithm:</span>**
   - This algorithm uses two passes of Depth-First Search (DFS). The first pass processes nodes in the original graph, and the second pass processes nodes in the transposed graph (with all edges reversed), in the order defined by the first pass.
   - The time complexity of Kosaraju's Algorithm is `O(V + E)`, where `V` is the number of vertices and `E` is the number of edges.

2. **<span style="color: #e67e22;">Tarjan's Algorithm:</span>**
   - This algorithm is based on DFS and uses a stack to track visited nodes. It identifies SCCs by detecting the roots of DFS trees.
   - Tarjan's Algorithm also runs in `O(V + E)` time.

3. **<span style="color: #8e44ad;">Gabow's Algorithm:</span>**
   - Similar to Tarjans, but uses two stacks to improve simplicity and possibly performance in certain cases.
   - It also has a time complexity of `O(V + E)`.

### <span style="color: darkorange;">Applications of SCCs</span>

Strongly Connected Components are essential in various fields such as:
- **<span style="color: #8e44ad;">Web Crawling:</span>** SCCs can identify clusters of interlinked pages.
- **<span style="color: #27ae60;">Circuit Design:</span>** In electronic circuits, SCCs can help identify regions that exhibit feedback.
- **<span style="color: #d35400;">Optimization Problems:</span>** SCCs are used in solving graph-related optimization problems like finding the strongly connected subgraph with the maximum total weight.

### <span style="color: darkorange;">Example</span>
Consider a directed graph where some nodes are mutually reachable from one another, forming SCCs. Nodes within an SCC can reach each other, but nodes in different SCCs may not have direct paths to each other.

- **<span style="color: #2980b9;">Example SCC 1:</span>** If nodes A, B, and C form an SCC, there are paths A  B, B  C, and C  A.
- **<span style="color: #e74c3c;">Example SCC 2:</span>** Nodes D, E, and F form another SCC with paths D  E, E  F, and F  D.

### <span style="color: darkorange;">DFS</span>

```python
def dfs(graph: Dict[int, List[int]], node: int, visited: Set[int], stack: List[int]) -> None:
    visited.add(node)
    for neighbor in graph[node]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited, stack)
    stack.append(node)
    
```

### <span style="color: darkorange;">Graph Transpose Function </span>

```python
def transpose(input_graph: Dict[int, List[int]]) -> Dict[int, List[int]]:
    t_graph: Dict[int, List[int]] = {node: [] for node in input_graph}
    for node in input_graph:
        for neigh in input_graph[node]:
            t_graph[neigh].append(node)
    return t_graph
```

### <span style="color: darkorange;">Kosaraju Function </span>

```python

def kosaraju(graph: Dict[int, List[int]]) -> List[Dict[int, List[int]]]:
    visited: Set[int] = set()
    stack: List[int] = []

    # Step 1: Fill the stack with vertices in the order of their finishing times
    for node in graph:
        if node not in visited:
            dfs(graph, node, visited, stack)

    # Step 2: Transpose the graph
    t_graph = transpose(graph)

    # Step 3: Process nodes in reverse order of their finishing times
    visited.clear()
    components: List[Dict[int, List[int]]] = []

    while stack:
        node = stack.pop()
        if node not in visited:
            component_nodes = []
            # Collect all nodes in the current SCC
            dfs(t_graph, node, visited, component_nodes)

            # Create the subgraph for this component
            component_graph: Dict[int, List[int]] = {n: [] for n in component_nodes}
            for n in component_nodes:
                component_graph[n] = [neigh for neigh in t_graph[n] if neigh in component_nodes]

            components.append(component_graph)
    
    return components
```

### <span style="color: darkorange;">DFS for Tarjan</span>

```python
def dfs_tarjan(graph: Dict[int, List[int]], node: int, index: Dict[int, int], lowlink: Dict[int, int],
               stack: List[int], on_stack: Dict[int, bool], current_index: int,
               components: List[Dict[int, List[int]]]) -> int:
    index[node] = current_index
    lowlink[node] = current_index
    current_index += 1
    stack.append(node)
    on_stack[node] = True

    for neigh in graph[node]:
        if index[neigh] == -1:  # If neighbor has not been visited
            current_index = dfs_tarjan(graph, neigh, index, lowlink, stack, on_stack, current_index, components)
            lowlink[node] = min(lowlink[node], lowlink[neigh])
        elif on_stack[neigh]:  # If neighbor is in the stack, it's part of the current SCC
            lowlink[node] = min(lowlink[node], index[neigh])

    # If node is a root node, pop the stack and generate an SCC
    if lowlink[node] == index[node]:
        scc = []
        while True:
            v = stack.pop()
            on_stack[v] = False
            scc.append(v)
            if v == node:
                break

        # Create the subgraph for the SCC
        component_graph = {n: [] for n in scc}
        for n in scc:
            component_graph[n] = [neigh for neigh in graph[n] if neigh in scc]
        components.append(component_graph)

    return current_index
```

### <span style="color: darkorange;">Tarjan Function </span>

```python
def tarjan(graph: Dict[int, List[int]]) -> List[Dict[int, List[int]]]:
    index = {node: -1 for node in graph}  # To store the index of each node
    lowlink = {node: -1 for node in graph}  # To store the lowest link value of each node
    on_stack = {node: False for node in graph}  # To check if a node is in the stack
    stack = []
    components: List[Dict[int, List[int]]] = []

    current_index = 0
    for node in graph:
        if index[node] == -1:
            current_index = dfs_tarjan(graph, node, index, lowlink, stack, on_stack, current_index, components)

    return components
```
